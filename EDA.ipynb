```python
event_hist_ct_unstack = event_hist_ct.unstack()
```

### Total Amt Spent on Fully Paid Claim, CAD Diabetes, Hypertension, CPD and CHF
### Export as .csv file

```python
history.head()
```

```python
history['event_descr'].unique()
```

```python
event_list = ['Fully Paid Claim', 'New diagnosis - CAD','New diagnosis - CPD',
              'New diagnosis - CHF', 'New diagnosis - Diabetes', 'New diagnosis - Hypertension']
```

```python
history_amt = history.loc[history['event_descr'].isin(event_list),['id','event_descr','event_attr3']]
```

```python
history_amt = history_amt.rename(columns={"event_attr3":"Amt_Spent"})
```

```python
history_amt.head()
```

```python
history_amt.dtypes
```

```python
history_amt.Amt_Spent = history_amt.Amt_Spent.astype(float)
```

```python
history_amt_agg = history_amt.groupby(['id','event_descr']).sum()
```

```python
history_amt_agg
```

```python
history_amt_agg_unstack = history_amt_agg.unstack()
```

```python
history_amt_agg_unstack.head()
```

```python
amt_agg_tbl = history_amt_agg_unstack.fillna(0).reset_index()
```

```python
amt_agg_tbl.head()
```

```python
amt_agg_tbl.to_csv('/Users/Janey_jiayan/Downloads/Humana_Data/amt_agg_tbl.csv')
```

### Average Amt Spent per Surgery 
### Export as .csv file

```python
history.head()
```

```python
history_amt_surgery = history.loc[history['event_descr']=='Surgery',['id','event_descr','event_attr3']]
```

```python
history_amt_surgery.head()
```

```python
history_amt_surgery = history_amt_surgery.rename(columns={"event_attr3":"Avg_Amt_Spent_Surgery"})
```

```python
history_amt_surgery.head()
```

```python
history_amt_surgery.dtypes
```

```python
history_amt_surgery.Avg_Amt_Spent_Surgery = history_amt_surgery.Avg_Amt_Spent_Surgery.astype(float)
```

```python
history_amt_surgery_avg = history_amt_surgery.groupby(['id','event_descr']).mean()
```

```python
history_amt_surgery_avg = history_amt_surgery_avg.unstack()
```

```python
history_amt_surgery_avg.head()
```

```python
surgery_amt_avg_tbl = history_amt_surgery_avg.fillna(0).reset_index()
```

```python
surgery_amt_avg_tbl.head()
```
